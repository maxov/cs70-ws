```julia
using Gadfly

x = 1:20
function poission(lambda)
    function dist(k)
        k = BigFloat(k)
        return exp(-lambda) * ((lambda ^ k) / factorial(k))
    end 
    return dist
end

function binom(n, p)
    p = BigFloat(p)
    function dist(k)
        k = BigFloat(k)
        return binomial(n, k) * ((1 - p) ^ (n - k)) * (p ^ k)
    end 
    return dist
end 
```

```julia
myp = poission(10)
myb = binom(100)
```

```julia
p = plot(layer(x=x, y=map(myp, x), Geom.point),
         layer(x=x, y=map))
```

```julia
img = PNG("poisson_plot.png", 6inch, 4inch)
draw(img, p)
```
